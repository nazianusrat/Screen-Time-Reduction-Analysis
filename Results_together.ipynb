```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data from your 55-study systematic review
data = {
    "Intervention": [
        "Behavioral / Time Restriction",
        "Structural / Device-Based",
        "School-Based",
        "Family-Based",
        "Physical Activity",
        "Mindfulness-Based",
        "AI / Tech-Based"
    ],
    "Effect_Size": [0.27, 0.45, 1.47, 0.72, 3.11, 0.41, 0.22],
    "Success_Rate": [50, 55, 90, 88, 95, 70, 60],
    "Number_Studies": [4, 8, 5, 6, 3, 4, 5],
    "Sustainability": [35, 45, 75, 80, 85, 70, 40]
}

df = pd.DataFrame(data)

# Create the bubble chart
fig, ax = plt.subplots(figsize=(14, 8))

scatter = ax.scatter(df['Effect_Size'], df['Success_Rate'],
                    s=df['Number_Studies']*500,
                    c=df['Sustainability'],
                    cmap='RdYlGn',
                    alpha=0.65,
                    edgecolors='black',
                    linewidth=0.5,
                    vmin=30, vmax=90)

# Add intervention labels to bubbles
for idx, row in df.iterrows():
    ax.annotate(row['Intervention'],
               xy=(row['Effect_Size'], row['Success_Rate']),
               fontsize=9,
               fontweight='2',
               ha='center',
               va='center')

# Colorbar for sustainability
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Sustainability Score (%)', fontsize=11, fontweight='bold')

ax.set_xlabel('Effect Size', fontsize=12, fontweight='bold')
ax.set_ylabel('Success Rate (%)', fontsize=12, fontweight='bold')
ax.set_title('Smartphone Screen Time Interventions: Effect Size vs Success Rate\nBubble size = Number of studies | Color = Sustainability',
             fontsize=13, fontweight='bold', pad=20)

ax.grid(True, alpha=0.3)
ax.set_xlim(-0.1, 3.3)
ax.set_ylim(40, 100)

# Add quadrant reference lines
ax.axvline(x=0.8, color='gray', linestyle=':', linewidth=2, alpha=0.5)
ax.axhline(y=75, color='gray', linestyle=':', linewidth=2, alpha=0.5)

plt.tight_layout()
plt.savefig('final_results_bubble_chart.png', dpi=300, bbox_inches='tight')
plt.show()


```
