```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(14, 8))

# Interventions with effectiveness and research gaps
interventions = ["Physical Activity", "School-Based", "Family-Based", "Behavioral",
                 "Device-Based", "Mindfulness", "Tech-Based", "Notifications",
                 "Tracking", "Nudges", "Awareness"]
effect_sizes = [3.11, 1.47, 0.72, 0.27, 0.45, 0.41, 0.22, 0.00, 0.00, 0.05, -0.10]
research_gaps = [40, 50, 45, 55, 60, 65, 70, 80, 75, 70, 80]  # Estimated gaps
colors_list = ['steelblue', 'steelblue', 'steelblue', 'steelblue', 'steelblue',
               'steelblue', 'steelblue', 'lightcoral', 'lightcoral', 'lightcoral', 'lightcoral']
sizes = [abs(e)*750 + 300 for e in effect_sizes]

scatter = ax.scatter(research_gaps, effect_sizes, s=sizes, c=colors_list,
                    alpha=0.8, edgecolors='black', linewidth=0.5)

# Add labels to bubbles
for i, txt in enumerate(interventions):
    ax.annotate(txt, (research_gaps[i], effect_sizes[i]), fontsize=9, ha='center', fontweight='2')

ax.set_xlabel("Research Gap / Evidence Need (%)", fontweight='bold', fontsize=11)
ax.set_ylabel("Effect Size (Effectiveness)", fontweight='bold', fontsize=11)
ax.set_title("Key Findings: Intervention Effectiveness vs. Research Gaps",
            fontweight='bold', fontsize=13)
ax.axhline(y=0, color='gray', linestyle='--', linewidth=1.5, alpha=0.7)
ax.grid(True, alpha=0.5, linestyle='--')

# Add quadrant labels
ax.text(45, 2.8, 'High Effectiveness\nLower Research Need', fontsize=10,
       bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.3), ha='center')
ax.text(75, 1.5, 'Moderate Effectiveness\nHigh Research Need', fontsize=10,
       bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.3), ha='center')
ax.text(86, -0.08, 'Failed Interventions\nHigh Research Need', fontsize=10,
       bbox=dict(boxstyle='round', facecolor='lightcoral', alpha=0.3), ha='center')

plt.tight_layout()
plt.show()
```
