```python
import matplotlib.pyplot as plt
import pandas as pd

# Summarised intervention-level data (failed / null results from literature)
data = {
    "Intervention": [
        "Notification-Based",
        "Passive Tracking/Monitoring",
        "Digital Nudges (App Limits)",
        "Awareness Cues/Pop-ups",
        "Meta-Level Evidence (Umbrella Review)",
        "Simple Goal-Setting (No Enforcement)"
    ],
    "Effect_Size": [0.00, 0.00, 0.05, -0.10, 0.00, 0.00]  # Null/negative effects from RCTs & meta-analyses
}

df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(11,6))
bars = plt.bar(df["Intervention"], df["Effect_Size"], color="lightcoral")

# Add value labels
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 0.03,
        f"{bar.get_height():.2f}",
        ha='center',
        fontsize=10
    )

plt.ylabel("Effect Size (Standardized)")
plt.xlabel("Intervention Type")
plt.title("Interventions That Failed to Reduce Smartphone Screen Time")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```
