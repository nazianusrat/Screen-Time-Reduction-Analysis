```python
import matplotlib.pyplot as plt
import pandas as pd

# Summarised intervention-level data
data = {
    "Intervention": [
        "Behavioral / Time Restriction",
        "Structural / Device-Based",
        "School-Based",
        "Family-Based",
        "Physical Activity",
        "Mindfulness-Based",
        "AI / Tech-Based"
    ],
    "Effect_Size": [0.27, 0.45, 1.47, 0.72, 3.11, 0.41, 0.22]
}

df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(11,6))
bars = plt.bar(df["Intervention"], df["Effect_Size"], color="steelblue")

# Add value labels
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 0.03,
        f"{bar.get_height():.2f}",
        ha='center',
        fontsize=10
    )

plt.ylabel("Effect Size (Standardized)")
plt.xlabel("Intervention Type")
plt.title("Effectiveness of Screen Time Reduction Interventions for Smartphone")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```
